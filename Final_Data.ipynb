```python
stock_names = db.raw_sql('select * from crsp_a_stock.dsfhdr')
display(stock_names)
```

```python
stock_names['enddat'] = pd.to_datetime(stock_names['enddat'])
stock_names['begdat'] = pd.to_datetime(stock_names['begdat'])
current_stock_names = stock_names[stock_names['enddat'] == pd.Timestamp('2023-12-29')]
current_stock_names = current_stock_names[current_stock_names['begdat'] <= pd.Timestamp('2019-01-01')].reset_index()

display(current_stock_names)
```

```python
db.describe_table(library='crsp_a_stock',table='dsf')
```

```python
def get_stock_data(sample_stock):
    cusip_list = sample_stock['cusip'].tolist()
    return tuple(cusip_list)

def query_data(cusip):
    prm = {'cus' : cusip, 'start_date' : '2019-01-01', 'none' : 'NaN'}
    data = db.raw_sql('select cusip, prc, openprc, date from crsp_a_stock.dsf where cusip == %(cus)s and date >= %(start_date)s', params = prm)
    return data

display(query_data('45383610'))
```

```python
columns_to_add = ['htick', 'hcomnam', 'hsiccd', 'numshr']

for col in columns_to_add:
    data[col] = None

last_cusip = None
for i in range(len(data)):
    if data['cusip'][i] == last_cusip:
        for col in columns_to_add:
            data.at[i, col] = data.at[i-1, col]
        continue
    for col in columns_to_add:
        data.at[i, col] = current_stock_names[current_stock_names['cusip'] == data['cusip'][i]][col].values[0]
    last_cusip = data['cusip'][i]

display(data)
print(data['htick'].unique())
```

```python
# Group the data by 'ticker'
grouped = data.groupby('htick')

# Save each group to a separate CSV file
for ticker, group in grouped:
    filename = f"./company_price/{ticker}.csv"
    group.to_csv(filename, index=False)
```
