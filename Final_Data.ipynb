```python
# output the csv file of the data with the given cusip, adding more information about the company
def output_csv(stock, data, years = 5):
    columns_to_add = ['htick', 'hcomnam', 'hsiccd', 'numshr']
    for col in columns_to_add:
        data[col] = data['cusip'].map(stock.set_index('cusip')[col])
        filename = f"./company_price/{data['htick'][0]}_{years}.csv"
        data.to_csv(filename, index = False)

# query the all data of stocks, and output the csv file, needs to run for over 30 minutes
def query_data(stocks, start_d = '2019-01-01', end_d = '2023-12-29'):
    for stock in stocks['cusip'].unique():
        prm = {'cus' : stock, 'start_date' : start_d, 'end_date' : end_d}
        data = db.raw_sql('select cusip, prc, openprc, date from crsp_a_stock.dsf where cusip = %(cus)s and date >= %(start_date)s and date <= %(end_date)s', params = prm)
        output_csv(stocks, data)

#query_data(current_stock_names)
```
