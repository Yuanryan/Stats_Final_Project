```python
# output the csv file of the data with the given cusip, adding more information about the company
def output_csv(stock, data, years = 5):
    columns_to_add = ['htick', 'hcomnam', 'hsiccd', 'numshr']
    for col in columns_to_add:
        data[col] = data['cusip'].map(stock.set_index('cusip')[col])
        filename = f"./company_price_10/{data['htick'][0]}_{years}.csv"
        data.to_csv(filename, index = False)

# query the all data of stocks, and output the csv file, needs to run for over 30 minutes
def query_data(stocks, start_d = '2019-01-01', end_d = '2023-12-29'):
    for stock in stocks['cusip'].unique():
        prm = {'cus' : stock, 'start_date' : start_d, 'end_date' : end_d}
        data = db.raw_sql('select cusip, prc, openprc, date from crsp_a_stock.dsf where cusip = %(cus)s and date >= %(start_date)s and date <= %(end_date)s', params = prm)
        output_csv(stocks, data, 10)

#query_data(current_stock_names)
```

```python
import os
import pandas as pd

company_price_5_dir = 'company_price_5'
company_price_10_dir = 'company_price_10'
output_dir = 'company_price_15'
files_5 = os.listdir(company_price_5_dir) 
files_10 = os.listdir(company_price_10_dir)
common_base_names = set(file[:-6] for file in files_5 if file.endswith('_5.csv')).intersection( file[:-7] for file in files_10 if file.endswith('_10.csv'))


# Process the common files
for base_name in common_base_names:
    file_5 = f"{base_name}_5.csv"
    file_10 = f"{base_name}_10.csv"
    
    df_5 = pd.read_csv(os.path.join(company_price_5_dir, file_5))
    df_10 = pd.read_csv(os.path.join(company_price_10_dir, file_10))
    
    combined = pd.concat([df_10, df_5]).drop_duplicates().reset_index(drop=True)
    combined.to_csv(os.path.join(output_dir, f"{base_name}.csv"), index=False)
    print(f"Combined file saved as {base_name}.csv")


```
